```
# partionBy() - It is used to partition large dataset into smaller filesbased on one or more multiple columns
```

```
data = [(1, 'sujeet', 'male', 'IT'),(2, 'ajeet', 'male', 'SDE'),(3, 'akanksha', 'female', 'SDE')]
schema = ['id','name','gender','dep']

df = spark.createDataFrame(data, schema)
```

```
df.show()
```

```
df.write.parquet('/FileStore/employees', mode='overwrite', partitionBy='dep')
```

```
spark.read.parquet('/FileStore/employees').show()
```

```
spark.read.parquet('/FileStore/employees/dep=SDE').show()
```

```
spark.read.parquet('/FileStore/employees/dep=IT').show()
```
